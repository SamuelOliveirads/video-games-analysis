### 3.1.1 Customers hypothesis

- **Age**: Games targeted at individuals aged 18-35 (prime gaming age) have higher overall sales compared to those targeted at other age groups.

- **Occupation**: Games favored by students have higher sales in regions with large student populations.
  
- **Income**: Games priced above the median price have higher sales in high-income regions.
  
- **Number of Children**: Family-oriented games have 20% higher sales in regions with a higher average number of children per family.
  
- **Location**: Games in Japanese languages have 15% higher sales in the Others markets compared to the North American market.

### 3.1.2 Game Genre hypothesis

- **Action**: Action games have a sales median that's 10% higher than RPG games in the European market.

- **Strategy**: Strategy games have 15% higher sales on PC platforms compared to sony platforms in the North American market.

- **Adventure**: Adventure games released on microsoft and sony consoles see a 10% increase in overall sales in the European market compared to those released on PC.

- **Simulation**: Simulation games have a sales average that's 20% higher in the Japan market than in the North American market.

### 3.1.3 Platform hypothesis

- **Console**: Games released for PlayStation consoles have 20% higher sales in Japan compared to games released for Xbox.

- **Mobile**: Mobile games account for 40% of total sales in the Europe market.
  
- **VR**: VR games have double the sales in regions where the median income is above $40,000 compared to regions below this median income.
